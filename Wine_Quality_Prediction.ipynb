```python
#separate the data and label
X = df.drop('quality',axis=1)

print(X)
```

```python
"""making quality value value is 1 or 0
if quality_value>=7
  print("Good quality wine")
else:
  print("Bad quality wine")
"""
Y = df['quality'].apply(lambda quality_value: 1 if quality_value>=7 else 0)
print(Y)
```

```python
#spliting training data and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
#model training
#random forest model
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy : ', test_data_accuracy)
```

```python
#predictive system

#input_data = (11.2,0.28,0.56,1.9,0.075,17,60,0.998,3.16,0.58,9.8)
input_data = (7.3,0.65,0,1.2,0.065,15,21,0.9946,3.39,0.47,10)

# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
